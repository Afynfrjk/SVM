```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
dir(iris)
```

```python
iris.target_names #OBJECTIVE IS: three types of target_names, we have to identify from feature_names
```

```python
iris.feature_names
```

```python
df=pd.DataFrame(iris.data, columns = iris.feature_names)
df.head(10)
```

```python
df["target"]=iris.target #0 is for 'setosa', 1 is for 'versicolor', 2 is for 'virginica'
df.head(150)
```

```python
df[df.target==1].head()
#df[df.target==0].head()
#df[df.target==2].head()
```

```python
df["flower_name"]=df.target.apply(lambda x: iris.target_names[x]) #apply new colum from target colum using apply function
df.head(120)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df0=df[df.target==0]
df1=df[df.target==1]
df2=df[df.target==2]
```

```python
df0.head()
#df1.head()
#df2.head()
```

```python
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(df0["sepal length (cm)"], df0["sepal width (cm)"], color='green', marker='^')
plt.scatter(df1["sepal length (cm)"], df1["sepal width (cm)"], color='magenta', marker='+')
plt.scatter(df2["sepal length (cm)"], df2["sepal width (cm)"], color='blue', marker='*')
```

```python
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.scatter(df0["petal length (cm)"], df0["petal width (cm)"], color='green', marker='^')
plt.scatter(df1["petal length (cm)"], df1["petal width (cm)"], color='magenta', marker='+')
plt.scatter(df2["petal length (cm)"], df2["petal width (cm)"], color='blue', marker='*')
```

**from above two plots, we can easiy identigy 'df0' (green points) which represents 'Setosa'**

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(['target','flower_name'], axis='columns')
x.head()
```

```python
y=df.target
y
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2)
print(len(x_train))
print(len(x_test))
print(len(y_train))
print(len(y_test))
```

```python
from sklearn.svm import SVC
model=SVC()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test, y_test)
```
